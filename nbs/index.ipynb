# Agent Imputer
 > Improving the Accuracy of Player Location Inference in Sports Matches Using Multi-Agent Spatial Imputation with Limited Observations.

## Overview

In this project, our primary aim is to leverage the research from the paper [Inferring Player Location in Sports Matches](https://arxiv.org/abs/2302.06569) to develop a model for analyzing soccer data. Specifically, we aim to adapt the multi-agent spatial imputation framework presented in the paper to our soccer data and generate a model that can accurately infer the location of players during a match.

To accomplish this goal, we will first conduct a thorough analysis of our soccer data to determine its characteristics, including the types of data that are available and the format in which they are presented. We will then explore the methodology presented in the paper and assess how it can be modified and applied to our soccer data.

Once we have adapted the model to our data, we will generate a wide training set and use it to train the model.

Finally, we will evaluate our model's performance by comparing its results with those generated by the paper. We will use a range of metrics to assess the accuracy of our model, including Mean absolute error.
